#Rater like model
Let's build a simple leak differential system

```python
#/###################/#
# Config
#

#ImportModules
import ShareYourSystem as SYS

#Debug Mode is True ?
SYS.DebugPrintBool=False

#/###################/#
# Build the model
#

#Define
MyLeaker=SYS.LeakerClass(
    ).mapSet(
        {
        '-Populations':{
            '|Default':{
                'LeakingUnitsInt':2,
                'LeakingMonitorIndexIntsList':[0,1],
                    '-Inputs':{
                        '|Krenel':{
                            'LeakingWeigthVariable':'#custom:#clock:200*ms:5.*mV*int(t==200*ms)',
                        },
                         '|Noise':{
                            'LeakingWeigthVariable':'#custom:#clock:1*ms:rand()*mV',
                        }
                    },
                '-Interactions':{
                        '|/':{
                            'LeakingWeigthVariable':-1.,
                        }
                    },
                }
            }
        }
    ).leak(
    )

#/###################/#
# Represent the model
#

#print
print('MyLeaker is ')
SYS._print(MyLeaker) 

```

```python
#/###################/#
# Simulate
#

MyLeaker.simulate(
    #SimulatingStopTimeFloat (ms here)
    500.
)

#/###################/#
# View
#

MyLeaker['/-Populations/|Default'].view(
    ).pyplot();
SYS.matplotlib.pyplot.show()
```
